## Background: The Cost of Olympic Sports

The Olympic games originated in 776 B.C. Greece and was revived centuries later in 1896, and has continued to this day. Four years after its revitalization, the modern ranking system of gold, silver, and bronze medals came into practice. Today, while the Olympic medals are proof of individual hard work and determination, the hefty financial commitment required from olympic athletes and the hosting country has transformed the competition into a testament of the wealth and power of participating nations. Consequently, the medal count of each country has become intertwined with national wealth.

With this association in mind, do countries with better economic status tend to produce athletes who obtain more medals in the Olympic games?

A glance at overall trends seems to suggest so: according to World Population Review, countries with a huge economic standing in the world have many Olympic medals - for example, the United States has a total of 2985 Olympic medals won since the first Olympic games while Zimbabwe has a total of 8 medals.

Existing research in this topic, however, indicates that the factors related to olympic performance are much more nuanced:
- One study <a name=\"cite_ref-1\"></a>[<sup>1</sup>](#cite_note-1) investigates the impact of economic, demographic, geographic, and social factors on Olympic performance using data from the 2016 Rio Olympics across 207 countries. Utilizing Quantile and Tobit Regression models, the research reveals that factors such as a country's inflation rate, economically active population, and income classification significantly influence Olympic medal rankings. Interestingly, the study finds that a country's GDP, corruption level, number of athletes, and topographical features do not significantly affect Olympic success, challenging some conventional assumptions about the determinants of sporting excellence on the global stage.
- Another study <a name=\"cite_ref-2\"></a>[<sup>1</sup>](#cite_note-2) examines how a country's economic conditions, including population size and political structures, impact its Olympic performance. Utilizing a fixed-effect Tobit regression model on data from the 1996, 2000, and 2004 Olympics, it finds that countries with larger populations and greater economic resources tend to win more medals. Additionally, the study explores the influence of political freedom and civil liberties on Olympic success, revealing complex relationships between these factors and a nation's medal tally.
Results from spending on olympic sports also seem to impact olympic performance diffently depending on the sport:
- A study on the 2016 Rio Olympics <a name=\"cite_ref-3\"></a>[<sup>1</sup>](#cite_note-3) additionally found that greater spending on swimming, track, and related sports by Australia, Canada, Japan, the U.S, and the U.K did yield to greater winnings, but the same effects were not as evident for team sports.
- According to Claire Farrow’s article, it states the costs and years of training for specific sports in the summer Olympics. For example, a Table Tennis athlete pays more than 25,000 dollars per year and more than four years of training to reach the level of an Olympic athlete. Another example is Tokyo Olympic sprint kayak athlete Shaye Hatchette: she spent more than 35,000 dollars in 2018 alone. Hatchette also says that people who do sports that aren’t “popular” have a hard time finding funding and sponsorships to support the cost of training.

### Dataset #1: Olympic results 

```
# Read in the Olympic_Results csv
olympic_results = pd.read_csv('olympic_results.csv')

# Take a look at the data
olympic_results.describe()
```

We can see that the original dataset contains observations for each country in each olympic event, organized by discipline (column 'slug_game') and event title (column 'event_title'). The columns we want to focus on are 'medal_types' and the country.
The study considered Olympic results from 2000 to 2020, focusing on the number and types of medals won. Data cleaning included constraining the data to Summer Olympics, and dropping columns we don't need.

```
# Cleaning: Drop irrelevant columns such as Athlete information, and others
olympic_results = olympic_results.drop(columns=['participant_type', 'athletes', 'rank_equal',
                                                'rank_position', 'athlete_url', 'athlete_full_name', 
                                                'value_unit', 'value_type', 'country_code'])

# Constrain data to summer olympics from the year 2020
olympic_results = olympic_results[olympic_results['slug_game'].str.contains('2020')]

# Take a looke at the cleaned data
olympic_results
```

The nans in the column "medal_type" mean no medal was received in that category. This is important, so we do not drop nans.

The 'slug_game' column encodes both the host city and year of the olympics in a string. We can separate this information into a new integer column that stores year values, and another string column that stores host cities to make it easier to handle the data.

```
# Functions to separate information encoded by 'slug_game'
# Extract string host city from input string
def get_city(string):
    string = string.lower()
    string = re.sub('-\d+',' ',string)
    string = string.strip()
    return string

# Extract year from input string
def get_year(string):
    string = string.lower()
    string = re.sub('\D+',' ',string)
    year = string.strip()
    return int(year)
```

```
# Apply functions to create a new column in dataset for host city
olympic_results['host_city']=olympic_results['slug_game'].apply(get_city)

# Create new column in dataset for olympic year
olympic_results['year']=olympic_results['slug_game'].apply(get_year)

# Drop 'slug_game' column
olympic_results=olympic_results.drop(labels='slug_game',axis=1)
```

To quantify the olympic performance of each country based on medal count, we assign each medal to a point value. Gold is worth 3 points, silver is worth 2 points, and bronze 1 point. We assign the point count to a new column in the table.

```
# Function that converts medal_type string to point number
def comp_points(medal):
    # 0 points when no medal is won

    if medal=='GOLD':
        return 3
    elif medal=='SILVER':
        return 2
    elif medal=='BRONZE':
        return 1
    else:
        return 0

# Apply function to dataframe
olympic_results['medal_points']=olympic_results['medal_type'].apply(comp_points)

# Drop 'medal_type' column
olympic_results=olympic_results.drop(labels='medal_type',axis=1)
```

## Bringing All the Data Together

Having the data all in one place will be easier to work with, but we are presently undecided. We also plan to ensure that the data from each of these datasets are consistent in type for ease of working with them during our analysis (i.e. making sure all numbers are ints or floats and such).
```
countries = olympic_results['country_name'].unique()

# Set up new dataframe where data will be combined
medal_perf2020 = pd.DataFrame(columns=['Country', 'Gold', 'Silver', 'Bronze', 'Points', 'Population', 'GDP'])

countries = olympic_results['country_name'].unique()

for country in countries:
    country_df = olympic_results[olympic_results['country_name'] == country]
    medal_counts = country_df['medal_points'].value_counts()

    # assign values to each columns
    quant_row = {'Country': country, 
                 'Gold': medal_counts.get(3, 0),
                 'Silver': medal_counts.get(2, 0),
                 'Bronze': medal_counts.get(1, 0),
                 'Points': (medal_counts.get(3, default=0) + medal_counts.get(2, default=0) + medal_counts.get(1, default=0)),
                 'Population': (population_2020[population_2020['Country'] == country]['Population (2020)'].iloc[0]),
                 'GDP': (country_gdp[country_gdp['Country Name']==country]['2020'])
                }
    medal_perf2020 = pd.concat([medal_perf2020, pd.DataFrame([quant_row])], ignore_index=True)

medal_perf2020["GDP"] = country_gdp["2020"]

medal_perf2020 = medal_perf2020.sort_values(by='Points', ascending=False).reset_index(drop=True)

medal_perf2020
```
